```python
#QQ plots
import scipy.stats as stats
def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)
    plt.title('Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('RM quantiles')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')
    plt.show()
```

```python
diagnostic_plots(data, 'Dexa_Freq_During_Rx')
```

```python
#Dexa_Freq_During_Rx is heavily right skewed. It appears that there are quite a few outliers at the right end of the distribution, as shown by the boxplot.
```

```python
# the following function determines the upper and lower limits for the skewed distribution
def find_skewed_boundaries(data, variable, distance):
    IQR = data[variable].quantile(0.75) - data[variable].quantile(0.25)
    lower_boundary = data[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = data[variable].quantile(0.75) + (IQR * distance)
    return upper_boundary, lower_boundary
```

```python
# determining outliers using the interquantile proximity rule ie (IQR * 1.5, the standard metric)
upper_boundary, lower_boundary = find_skewed_boundaries(data, 'Dexa_Freq_During_Rx', 1.5)
upper_boundary, lower_boundary
```

```python
#looking at the number and percentage of outliers for LSTAT
print('total umber of patients: {}'.format(len(data)))
print('Patients with Dexa_Freq_During_Rx bigger than 32: {}'.format(
    len(data[data['Dexa_Freq_During_Rx'] > upper_boundary])))
print()
print('%Patients with Dexa_Freq_During_Rx bigger than 32: {}'.format(
    len(data[data['Dexa_Freq_During_Rx'] > upper_boundary])/len(data)))
```

```python
def find_skewed_boundaries(df, variable, distance):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)
    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)
    return upper_boundary, lower_boundary
```

```python
# limits for CRIM
CRIM_upper_limit, CRIM_lower_limit = find_skewed_boundaries(data, 'Dexa_Freq_During_Rx', 1.5)
CRIM_upper_limit, CRIM_lower_limit
```

```python
# highlighting outliers in the data set
outliers_CRIM = np.where(data['Dexa_Freq_During_Rx'] > CRIM_upper_limit, True,
                       np.where(data['Dexa_Freq_During_Rx'] < CRIM_lower_limit, True, False))
```

```python
# trimming our dataset by removing outliers
data_trimmed = data.loc[~(outliers_CRIM), ]
data.shape, data_trimmed.shape
```

```python
diagnostic_plots(data_trimmed, 'Dexa_Freq_During_Rx')
```
